```python
##### import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
#plt.style.use('seaborn-pastel')


fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-1.5, 1.5))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Sine Wave Signal')
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                                   frames=200, interval=40, blit=True)

plt.tight_layout()
anim.save('sine_wave.gif', writer='imagemagick')
plt.show()
```

```python
import numpy
from numpy import pi


t, sample_rate, frequency, phase = numpy.linspace(0, 2, 1001), 10, 1, 0
signal = lambda t, f=1., phi=0: numpy.sin(2*pi*t*f-phi)

fig, ax = plt.subplots()
fig.set_size_inches(8,4)

#@interact(sample_rate=(1,25, 0.1), frequency=(-9,10, 0.1), phase=(-6, 6, 0.1))
def update(sample_rate=sample_rate, frequency=frequency, phase=0.):
    ax.cla()
    t_samples = numpy.arange(min(t), max(t), 1./sample_rate)
    x, x_sampled = signal(t, frequency, phase), signal(t_samples, frequency, phase)
    ax.plot(t, x, label='Original Signal')
    ax.stem(t_samples, x_sampled, linefmt='r', label='Sampled', use_line_collection=True)
    ax.plot(t_samples, x_sampled, '--k', label='Recovered')
    ax.legend(loc='lower left')
    fig.canvas.draw_idle()
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Sine Wave Signal')
    fig.tight_layout()

    
def animate(frame):
    phase = 2*pi*frame/200
    update(phase=phase)

anim = FuncAnimation(fig, animate, frames=200, interval=30, blit=True)

plt.tight_layout()
anim.save('sine_wave_sampled.gif', writer='imagemagick')
plt.show()
```
